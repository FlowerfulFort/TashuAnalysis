```python
import pandas as pd
import os
from typing import List
file_format = '2022_{0:02d}daejeon_.csv'
div = {
    '계': '',
    '남': '.1',
    '여': '.2'
}
df: List[pd.DataFrame] = []

def str2numeric(target: pd.Series) -> pd.Series:
    return target.apply(lambda x: int(''.join(x.split(','))))

for i in range(1, 13):
    filename = file_format.format(i)
    temp = pd.read_csv(os.path.join('data', filename), dtype=str)
    temp.rename(columns={'Unnamed: 0': '지역'}, inplace=True)
    temp.loc[:, '총인구수': '100세 이상'] = temp.loc[:, '총인구수':'100세 이상'].apply(str2numeric)
    df.append(temp)

df[0].head()
```

```python
new_df: List[pd.DataFrame] = []
for i in range(len(df)):
    new_df.append(df[i].loc[:, '지역':'100세 이상'])
new_df[0].head()
```

```python
mydf = pd.DataFrame(columns=new_df[0].columns)
mydf
```

```python
for i in range(len(new_df)):
    mydf = pd.concat([mydf, pd.DataFrame([new_df[i].loc[0]])], ignore_index=True)
mydf.rename(columns={'지역':'월'}, inplace=True)
for i in range(len(mydf['월'])):
    mydf['월'][i] = '{}월'.format(i+1)
mydf
```

```python
mydf.to_csv(os.path.join('preprocessed', 'population_by_month.csv'))
```
