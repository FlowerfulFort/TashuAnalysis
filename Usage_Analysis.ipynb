```python
from matplotlib import pyplot as plt
import folium

pos_cnu = [36.3687979, 127.3468559]
map = folium.Map(location=pos_cnu, tiles='OpenStreetMap', zoom_start=13)
for _, data in df_rental.iterrows():
    folium.Marker(location=[data['위도'], data['경도']],
                  popup=folium.Popup(html=folium.IFrame(data['대여소명'], height=60), 
                                     min_width=200, max_width=200),
                  icon=folium.Icon()
    ).add_to(map)
map
```
