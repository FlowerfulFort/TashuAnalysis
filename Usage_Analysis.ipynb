```python
rental_cnu = [1016, 369, 1015, 366, 368, 1014, 367]

origin_data_path = os.path.join('preprocessed', '2022_season2_processed.csv')
origin_data = pd.read_csv(origin_data_path,
                          index_col=0,
                          parse_dates=['대여일시', '반납일시'],
                          infer_datetime_format=True)

origin_data.head()
```

```python
import numpy as np
def retain_from_index_array(target: pd.Series, index: np.ndarray) -> pd.Series:
    c = target.copy()
    for i, v in target.items():
        if v in index:      # if index array contains v
            c[i] = True     # retain this row.
        else:               # if not contains v
            c[i] = False    # remove this row.
    return c
filtered = origin_data.loc[retain_from_index_array(origin_data['반납스테이션'], rental_cnu)]
# def stamp2date(ts) -> str:
#     return ''.join([ts.year,ts.month,ts.day])
# df.apply
```

```python
stat_dict = {}

def addToDict(d: dict, station: int):
    if not str(station) in d.keys():
        d[str(station)] = 1
    else:
        d[str(station)] += 1

filtered['대여스테이션'].apply(lambda x: addToDict(stat_dict, x))
stat_dict
```

```python
x = stat_dict.keys()
y = stat_dict.values()

rentaldf = pd.DataFrame(data={'대여스테이션': x, '빈도': y})
rentaldf
```

```python
rentaldf = rentaldf.sort_values(by='빈도', ascending=False)
top10rental = rentaldf.head(30)
top10rental
```

```python
# 대여(-)/음봉, 반납(+)/양봉
target_rental = stat_dict.keys()

def addToDict(d, date:str, val: int):
    if not date in d.keys():
        d[date] = val
    else:
        d[date] += val

Average = {}
for i in target_rental:
    ii = int(i)

    # 하루에 반납-대여 차이가 얼마나 나는지 기록하는 딕셔너리
    # key = YYYYmmdd | value = (반납-대여)
    moment_temp = {}

    rent_df= origin_data.loc[origin_data['대여스테이션'] == ii]
    return_df = origin_data.loc[origin_data['반납스테이션'] == ii]
    rent_temp = rent_df['대여일시']
    return_temp = return_df['반납일시']

    # 대여했으므로 스테이션 잔여 오프셋에서 -1
    for d in rent_temp:
        stringified = f'{d.year}{d.month:02d}{d.day:02d}'
        addToDict(moment_temp, stringified, -1)
    # 반납했으므로 +1
    for d in return_temp:
        stringified = f'{d.year}{d.month:02d}{d.day:02d}'
        addToDict(moment_temp, stringified, 1)

    # 평균계산
    Average[i] = sum(moment_temp.values())/len(moment_temp)

for r in top10rental['대여스테이션']:
    print(r, Average[r])
```

```python
from math import atan
map3 = folium.Map(location=pos_cnu, tiles='OpenStreetMap', zoom_start=13)
fg_marker = folium.FeatureGroup(name='Location_marker').add_to(map3)
fg_chart = folium.FeatureGroup(name='Usage').add_to(map3)
line_offset = 0.0007    # 중심점으로부터 차트를 얼마만큼 떨어진 거리에서 보여줄지 결정

def scaler(x):  # 차트 스케일 함수
    return 0.002*atan(x)

for st in top10rental['대여스테이션']:
    # data = rental_origin_data.loc[rental_origin_data['번호']==int(st)]
    data = rental_origin.loc[rental_origin['번호'] == int(st)]

    # 위치를 보기 쉽게 큰 마커로 표현해줌
    folium.Marker(location=[data['위도'], data['경도']],
                  popup=folium.Popup(html=folium.IFrame(f"{data.iloc[0]['대여소명']}[{st}]: {stat_dict[st]}", height=60), 
                                     min_width=200, max_width=200),
                  icon=folium.Icon()
    ).add_to(fg_marker)

    # 중심점으로부터 차트를 그릴 곳까지 잇는 선
    loc = [(float(data['위도']), float(data['경도'])), (float(data['위도']), float(data['경도'])+line_offset)]
    folium.PolyLine(locations=loc, color='orange', weight=5).add_to(fg_chart)
    
    # 중심점 원
    folium.Circle(location=[data['위도'], data['경도']],
                  radius=20,
                  color='#eb920c',
                  fill_color='#000000',
                  popup=folium.Popup(html=folium.IFrame(f"{data.iloc[0]['대여소명']}[{st}]: {stat_dict[st]}", height=60), 
                                     min_width=200, max_width=200)
    ).add_to(fg_chart)

    # 차트
    chart_offset = Average[st]    
    folium.PolyLine(locations=(loc[1], (loc[1][0]+scaler(chart_offset), loc[1][1])),
                    color='red' if chart_offset < 0 else 'blue',
                    weight=5).add_to(fg_chart)

folium.LayerControl(collapsed=False).add_to(map3)
map3
```
